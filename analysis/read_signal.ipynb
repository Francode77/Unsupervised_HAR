```python
# This file reads any .csv file in the folder /data/unlabeled and plots graphs for x

import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
from os import listdir
import os

%matplotlib inline

path=os.path.join('../data/unlabeled')
list_of_files=listdir(path)
colors=['#85C34D','#B593CD','#C5C34D']
lwidth=[.2,.2,2]
x=0

def plot(df,file_id,y_data): 
    global colors
    global x
    print (file_id)
    sns.set_theme(style='whitegrid')
    sns.set(rc={'figure.figsize':(16,6)}) 
    sns.lineplot(
        data=df, x=df.index, y=y_data, label=y_data, 
        linewidth=lwidth[x], linestyle='solid', color=colors[x]
    ).set(title=f'{y_data} for {file_id}')
    plt.show() 
    x+=1
    if x>2:
        x=0

for file in list_of_files:
    if '.csv' in file:
        df=pd.read_csv(file)
        plot (df,file,'acc_x_r')
        plot (df,file,'gyr_x_r')
        plot (df,file,'kal_status_n')
        
```
